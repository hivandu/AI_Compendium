```python
import matplotlib.pyplot as plt
from  PIL import Image
```

```python
import numpy as np
image = np.array([
    [10, 10, 10, -3, -3, -3],
    [10, 10, 10, -3, -3, -3],
    [10, 10, 10, -3, -3, -3],
    [10, 10, 10, -3, -3, -3],
    [10, 10, 10, -3, -3, -3],
    [10, 10, 10, -3, -3, -3],
])
plt.imshow(image)
plt.show()
```

```python
def single_conv(image, filter):
    h = filter.shape[0]
    w = filter.shape[1]

    r_height = image.shape[0] - h+1
    r_width = image.shape[1] - w+1

    result = np.zeros(shape=(r_height, r_width))

    for i in range(image.shape[0] - h+1):
        for j in range(image.shape[1] - w+1):
            window = image[i: i+h, j: j+w]

            result[i][j] = np.sum(filter * window)

    return result

def conv(image, filters):
    results = [single_conv(image, f) for f in filters]
    return results

h_filter = np.array([
    [1, 0, -1],
    [1, 0, -1],
    [1, 0, -1]
])

w_filter = np.array([
    [1, 1, 1],
    [0, 0, 0],
    [-1, -1, -1]
])

myself = Image.open('./assets/chaheng.jpg').convert('L')

image = np.array(myself)

results = conv(image, [h_filter, w_filter])

plt.subplot(1, 3, 1)
plt.imshow(image)
plt.subplot(1, 3, 2)
plt.imshow(results[0])
plt.subplot(1, 3, 3)
plt.imshow(results[1])

plt.show()
```
