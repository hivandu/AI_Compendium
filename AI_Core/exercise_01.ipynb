# EXERCISE 01. Logistic Regression for Predicting Heart Disease

```python
# read Dataset
import pandas as pd
import os

data_path = os.environ.get('DATA_PATH')

data_file = data_path + 'AI_Cheats/heart.csv'
train_data = pd.read_csv(data_file)
```

```python
# Print a brief summary of the dataset
train_data.info()
```

### Meaning of Fields in the Dataset

| Field Name | Meaning |
|---|---|
| age | Age |
| sex | Gender (1 = Male, 0 = Female) |
| cp | Chest pain type (Value 1: Typical angina, Value 2: Atypical angina, Value 3: Non-anginal pain, Value 4: Asymptomatic) |
| trestbps | Blood pressure |
| chol | Cholesterol |
| fbs | Fasting blood sugar (> 120 mg/dl, 1 = True; 0 = False) |
| restecg | ECG results (0 = Normal, 1 = ST-T wave abnormality, 2 = Possible or definite left ventricular hypertrophy according to Estes' criteria) |
| thalach | Maximum heart rate |
| exang | Exercise induced angina (1 = Yes; 0 = No) |
| oldpeak | ST depression induced by exercise relative to rest |
| slope | Slope of the ST segment (Value 1: Upsloping, Value 2: Flat, Value 3: Downsloping) |
| ca | Number of vessels colored by fluoroscopy |
| thal | Type of defect (3 = Normal; 6 = Fixed defect; 7 = Reversable defect) |
| target | Disease presence (0 = No, 1 = Yes) |

```python
# Get the number of diseased and non-diseased
train_data.target.value_counts()
```

```python
# Separate gender into two columns, using True and False to indicate
sex = pd.get_dummies(train_data['sex'], prefix = 'sex')
train_data = pd.concat([train_data, sex], axis = 1)
train_data = train_data.drop(columns = ['sex'])
train_data.head(5)
```

```python
# Getting sample labels
y_data = train_data.target.values
train_data.shape
```

```python
# Get the sample feature set
x_data = train_data.drop(['target'], axis = 1)
x_data.shape
```

```python
# Delineate data sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size = 0.3, random_state=33)
```

```python
# Normalization
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
```

```python
# Training model parameters
standardScaler.fit(X_train)
```

```python
# Normalize training and test sets
X_train = standardScaler.transform(X_train)
X_test = standardScaler.transform(X_test)
```

```python
# Define the logistic regression model
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
# Calculate training set scores
log_reg.score(X_train, y_train)
```

```python
# Calculate test set scores
log_reg.score(X_test, y_test)
```

```python
# Text reports showing key disaggregated indicators
from sklearn.metrics import  classification_report
y_predict_log = log_reg.predict(X_test)
print(classification_report(y_test, y_predict_log))
```
