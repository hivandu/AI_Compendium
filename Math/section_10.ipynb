1. Iterative method for solving the minimum of a quadratic function

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from IPython import display
```

```python
# Define quadratic function
def f(x):
    y = x**2
    return y
```

```python
# Define corresponding derivative
def d(x):
    dy = 2*x
    return dy
```

```python
# User inputs initial value of x
x = float(input('Please enter the initial value of x: '))

# Define learning rate
lr = 0.9

# Prepare data points for the quadratic function
x_temp = np.linspace(-x, x, 1000)
y_temp = x_temp**2

while abs(x) > 0.01:
    # Clear previous output
    display.clear_output(wait=True)
    # Clear previous plot
    plt.clf()
    plt.plot(x_temp, y_temp)
    plt.scatter(x, f(x), s=30, c='red')
    plt.show()
    print('Current x value: ', x)
    plt.pause(0.7)
    dy = d(x)
    x -= lr * dy
else:
    display.clear_output(wait=True)
    plt.clf()
    plt.plot(x_temp, y_temp)
    plt.scatter(x, f(x), s=30, c='red')
    plt.show()
    print('Final x value: ', x)
```
