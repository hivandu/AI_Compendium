# Matploylib and Seaborn

```python
import numpy as np
import pandas as pd
from pylab import mpl
import matplotlib.pyplot as plt
plt.style.use('default')

import seaborn as sns
from matplotlib.font_manager import FontProperties

import os

dp = os.environ.get('DATA_PATH')
```

```python
# Data preparation
N = 500
x = np.random.randn(N)
y = np.random.randn(N)

# Scatter plot with Matplotlib
plt.scatter(x, y, marker='x')
plt.show()
```

```python
# Scatter plot with Seaborn
df = pd.DataFrame({'x':x, 'y':y})
sns.jointplot(x='x', y='y', data=df, kind='scatter')
plt.show()
```

## Radar Chart

```python
# Data preparation
labels = np.array(['Pushing', 'KDA', 'Survival', 'Teamfight', 'Growth', 'Output'])
stats = [76, 58, 67, 97, 86, 58]

# Prepare data for plot: angles, stats
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False)
stats = np.concatenate((stats, [stats[0]]))
angles = np.concatenate((angles, [angles[0]]))
labels = np.concatenate((labels, [labels[0]]))

# Check if angles and labels match
print(len(angles), len(labels))  

# Plot radar chart with Matplotlib
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)
ax.plot(angles, stats, 'o-', linewidth=2)
ax.fill(angles, stats, alpha=0.25)

ax.set_thetagrids(angles * 180 / np.pi, labels)
plt.show()
```

## Binary distribution map

```python
# Load data
flights = sns.load_dataset('flights')
flights.info()
flights
```

```python
# Bivariate distribution plots with Seaborn (scatter, KDE, Hexbin)
sns.jointplot(x='year', y='passengers', data=flights, kind='scatter')
sns.jointplot(x='year', y='passengers', data=flights, kind='kde')
sns.jointplot(x='year', y='passengers', data=flights, kind='hex')
plt.show()
```

```python
# View the distribution of passengers (univariate distribution)
sns.kdeplot(flights['passengers'], color='red', fill=True)
plt.show()
```

```python
# Seaborn draws pairs
sns.pairplot(flights)
plt.show()
```

## subplot

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0, 5, 0.1)
t2 = np.arange(0, 5, 0.02)

plt.figure()
plt.subplot(221)
plt.plot(t1, f(t1), 'r-')
plt.subplot(222)
plt.plot(t2, np.cos(2*np.pi*t2), 'r-')
plt.subplot(212)
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```
