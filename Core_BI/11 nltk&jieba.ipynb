# Word Cloud

```python
from wordcloud import WordCloud
import pandas as pd
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize

import os

dp = os.environ.get('DATA_PATH')
```

```python
# Remove Stop Words
def remove_stop_words(f):
    stop_words = ['Movie']
    for stop_word in stop_words:
        f = f.replace(stop_word, '')
    return f

# Generate Word Cloud
def create_word_cloud(f):
    print('Generating word cloud based on word frequency!')
    # Remove stop words (frequent words like "the", "of", "a")
    f = remove_stop_words(f)    
    cut_text = word_tokenize(f)

    # Uncomment to view tokenized words
    # print(cut_text)

    cut_text = ' '.join(cut_text)
    wc = WordCloud(
        max_words=100,
        width=1200,
        height=1000
    )
    wordcloud = wc.generate(cut_text)
    # Save the word cloud image
    wordcloud.to_file(dp + 'AI_Cheats/assets/wordcloud.jpg')
    # Display the word cloud image
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
```

```python
# Load data
data = pd.read_csv(dp + 'AI_Cheats/wordcloud/movies.csv')
data.head()
```

```python
# Read the title and genres Fields
title = ' '.join(data['title'])
genres = ' '.join(data['genres'])
all_word = title + genres
# Generate Word Cloud
create_word_cloud(all_word)
```

## Market_Basket_Optimisation

```python
# Load data
data = pd.read_csv(dp + 'AI_Cheats/wordcloud/Market_Basket_Optimisation.csv', header=None)
data.head()
data.values
```

```python
all_word = ''
temp_list = data.values
for temp in temp_list:
    for x in temp:
        if str(x) != 'nan': 
            all_word = all_word + ' ' + x

# Generate Word Cloud
create_word_cloud(all_word)
```
