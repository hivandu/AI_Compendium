```python
# Sales data comparison
y1 = [20,10,10,20,40,30,40,40,50,60,50,40]
y2 = [10,30,20,30,40,20,10,30,30,80,30,20]
x = range(1, 13)

# Set figure
plt.figure(figsize=(20, 8), dpi=80)

plt.plot(x, y1, color='orange', label="Hivan")
plt.plot(x, y2, color='green', label='colleague')

# Set x-axis ticks
xtick_labels = ['{} Mon'.format(i) for i in x]

plt.xticks(x, xtick_labels, rotation=45)

# Draw grid
plt.grid(alpha=0.4) # Grid can also have styles; here, the transparency is 0.4

# Add legend
# Set position loc: upper left, lower left, center left, upper center
plt.legend(loc='upper right')

plt.show()
```

```python
# Multiple subplots
x = np.arange(1, 100)
fig = plt.figure(figsize=(20, 10), dpi=80)

# Subplot 1
ax1 = fig.add_subplot(2,2,1)
ax1.plot(x, x)
# Subplot 2
ax2 = fig.add_subplot(2,2,2)
ax2.plot(x, x**2)
ax2.grid(color='r', linestyle='--', linewidth=1, alpha=0.3)
# Subplot 3
ax3 = fig.add_subplot(2,2,3)
ax3.plot(x, np.log(x))

plt.show()
```

```python
# Change the default display of axis
y = range(0, 14, 2)
x = [-3, -2, -1, 0, 1, 2, 3]
plt.figure(figsize=(20,8), dpi=80)

# Get current plot's axes
ax = plt.gca()

# Set the borders of the graph
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('blue')
ax.spines['left'].set_color('red')

# Set the position of the bottom border to the y-axis 0 position
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

plt.plot(x, y)
plt.show()
```
